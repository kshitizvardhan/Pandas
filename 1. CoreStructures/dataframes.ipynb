# DataFrame Parameters: The Blueprint
- Syntax: pd.DataFrame(data=None, index=None, columns=None, dtype=None)
- Instead of a name parameter (which a Series has), a DataFrame has a columns parameter to name its multiple Series.

```python
raw_data = [
    [25, 55000, "HR"],
    [38, 89000, "IT"],
    [22, 41000, "Finance"]
]

employee_df = pd.DataFrame(
    data=raw_data,
    index=["Emp01", "Emp02", "Emp03"],
    columns=["Age", "Salary", "Department"],
    dtype="object"
)

print(employee_df)
```

### Accessing Data
- DataFrame is 2-dimensional, accessing data is slightly different than a Series. 
- You have to think in terms of Columns, Rows, or specific Cells.

1. Selecting Columns (Extracting Features)
    - To grab a single column, use brackets with the column name. This returns a Pandas Series. 
    - To grab multiple columns, you must pass a list of column names inside the brackets. This returns a smaller DataFrame.

2. Selecting Rows and Cells (.loc and .iloc)
    - .loc is label-based, meaning you have to specify the name of the row or column you want to grab.
    - .iloc is integer position-based, meaning you have to specify the integer index of the row
    - Same as in Series, but now we use a coordinate system: [row, column].

```python
ages = employee_df["Age"] # Returns a Panda Series containing the ages of the employees
print(ages)
subset = employee_df[["Age", "Salary"]] # Returns a smaller DataFrame containing both the Age and Salary columns
print("\n")
print(subset)
```

```python
# Grab a entire row of data for a specific employee using loc
emp_2_data = employee_df.loc["Emp02"] # Accessing the row for Emp02 using loc, this will return a Series containing the data for that employee
print(emp_2_data)

# Grab a specific cell using .loc[row_label, column_label]
emp_2_salary = employee_df.loc["Emp02", "Salary"] # Accessing the Salary for Emp02 using loc, this will return the value of that cell
print("\nSalary for Emp02: ", emp_2_salary)

# 3. Grab a specific cell using .iloc[row_position, col_position]
# Row 0 (first row), Column 1 (second column -> Salary)
emp_2_salary = employee_df.iloc[1, 1]
print("Salary for Emp02: ", emp_2_salary)
```

### Creating new Column
- Can create a column simply by declaring it and assigning it a value (or a vectorized calc)

```python
# Giving everyone a $5000 bonus
employee_df["Total"] = employee_df["Salary"] + 5000
print(employee_df)
```

### Dropping rows or Cols
- Using .drop(), must specify the axis
    - axis=0 Refers to Rows (the default)
    - axis=1 Refers to Columns

```python
employee_df = employee_df.drop("Salary", axis=1) 
print(employee_df)
employee_df = employee_df.drop("Emp03", axis=0)
print(employee_df)
```

### Handling Missing Data
- When you have a DataFrame with NaN (missing) values, you have two main choices: drop them, or fill them (imputation).
    - df.dropna(axis=0): Drops any row that contains at least one missing value.
    - df.dropna(axis=1): Drops any column that contains at least one missing value.
    - df.fillna(value): Fills all missing values in the DataFrame with a specific value.

### Adding a single row
- Using .loc (Best for adding a single row by index label)

```python
employee_df.loc["Emp03"] = [30, "Marketing", 60000]
print(employee_df)
```

- pd.contact() [Preferred, Robust way]
    - If you are adding data on the fly and just want it to attach to the bottom (without worrying about specific index names), 
    - You convert the new row into a mini-DataFrame and concatenate them

```python
# The new data we want to add
new_employee_df = pd.DataFrame(
    data=[[28, "Sales", 45000]],
    index=["Emp04"],
    columns=["Age", "Department", "Total"]  # if we don't specify the columns, it will just add the new row with default column names like 0, 1, 2... which we don't want
)

# Concatenating the old df and the new row
employee_df = pd.concat([employee_df, new_employee_df])
print(employee_df)
```

### Adding Two Dataframes together

```python
df1 = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

df2 = pd.DataFrame({
    "A": [7, 8, 9],
    "B": [10, 11, 12]
})

combined_df = pd.concat([df1, df2], axis=0, ignore_index=True) # axis=0 means we want to concatenate along the rows, ignore_index=True means we want to reset the index after concatenation
print(combined_df)

combined_df = pd.concat([df1, df2], axis=1, ignore_index=True) # axis=1 means we want to concatenate along the columns, ignore_index=True means we want to reset the column names after concatenation
print(combined_df)
```
