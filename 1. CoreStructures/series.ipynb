# Series

```python
import pandas as pd
```

```python
ages_list = [22, 25, 27, 21, 23] # A Standard Python List of ages
ages_series = pd.Series(ages_list) # Convert the list to a Pandas Series
print(ages_series) # Print the Series

print(type(ages_series)) # Panda Series
```

- Pandas automatically assigned an index (0,1,2,3,4) on the left to our values on the right.
- It also tells us the data type (dtype: int64, which means a 64-bit integer)

### Custom Indexing, datatype and name
- Syntax: pd.Series(data=None, index=None, dtype=None, name=None)

1. data: The actual values (can be a list, dictionary, or even a single number)
2. index: Custom labels for each row (must match the length of the data)
3. dtype: Forcing a specific data type (e.g., floating point numbers)
4. name: Giving the Series a name (this becomes the column name in a DataFrame)

```python
month_data = pd.Series(
    data=[31, 28, 30, 31, 30, 31, 31, 30, 31, 30, 31, 31],
    index=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    dtype='float32',
    name='Days in Month'
)

print(month_data)
```

```python
dict = {"name": "Kshitiz", "age": 21, "subjects": ["Math" , "Physics", "Chemistry"], "is_student": True }  
dict_Series = pd.Series(dict)
print(dict_Series)
print(dict_Series.iloc[0]) # Accessing the first value using integer position with iloc
print(dict_Series.loc['name']) # Accessing the value using the index label with loc
```

```python
s = pd.Series([1, 2, 3])
s2 = pd.Series([10, 20, 30, 40, 50])
print(s + s2) # Element-wise addition of two Series, this will add the values at the same index together, 
# if the indexes are different it will return NaN for those indexes
# In NumPy, this would have resulted in an error because the shapes of the arrays are different, 
# But in Pandas it handles this gracefully by aligning the indexes and returning NaN for any indexes that don't match.
```

### Accessing Data
- In standard Python lists, you access data by its position (e.g., my_list[0]). 
- Because a Pandas Series has custom index labels, it has two distinct ways to grab data. Understanding the difference between these two is critical.

- .loc[] (Location): Looks up data by the label (the name of the index).
- .iloc[] (Integer Location): Looks up data by the position (the physical row number, starting at 0).

- Even the standard python method works, but not recommended.

```python
# Standard Python ways of accessing data
print(month_data['Feb']) # Accessing a single value using the index label, this still works
# print(month_data[1]) # Accessing a single value using the integer position. throws an error because the Series has a custom index otherwise it would work

# Pandas ways of accessing data
print(month_data.iloc[1]) # Accessing a single value using the integer position with iloc
print(month_data.loc['Feb']) # Accessing a single value using the index label with loc
```

### Vectorized Operations

```python
prices = pd.Series([10,20,30])

# Vectorized operations
multiply_prices = prices * 2
subtract_prices = prices - 5
add_prices = prices + 10
divided_prices = prices / 3
powered_prices = prices ** 2
floor_divided_prices = prices // 3
modulous_prices = prices % 2

print(multiply_prices)
print(subtract_prices)
print(add_prices)
print(divided_prices)
print(powered_prices)
print(floor_divided_prices) 
print(modulous_prices)
```

### Masking

```python
ages = pd.Series([18,15,20,30,89,3])
adults = ages >= 18
print(adults)  # Masks the Series with a boolean Series
print(ages[adults]) # Filters the Series using the boolean Series, this is called boolean indexing or masking, returns only the values that are True in the boolean Series
```

### Essential ML Preparation Methods

- series.isnull(): Returns True for missing data (NaN).

- series.fillna(value): Replaces missing data with a specific value.

- series.describe(): Generates summary statistics (mean, min, max, standard deviation) instantly.

- series.value_counts(): Counts the frequency of unique values (crucial for classification tasks).

```python
# Create a Panda Series from raw_scores = [88.0, 92.5, None, 105.0, 79.0]
# Note: None represents a missing value where the model crashed. 105.0 is clearly an impossible outlier since the max score is 100.

# Follow these steps to clean and transform the data:
# Create a Pandas Series from raw_scores and give it custom index labels: ['run1', 'run2', 'run3', 'run4', 'run5'].
# Fill that missing None value (which Pandas will convert to NaN) with a baseline score of 80.0.
# Filter the Series to remove that impossible outlier (keep only the scores that are <= 100).
# Transform the remaining scores by adding a 2.0 point bonus to every run using vectorized math.
```

```python
raw_scores = [88.0, 92.5, None, 105.0, 79.0]
raw_scores_series = pd.Series(
    data=raw_scores,
    index=['run1', 'run2', 'run3', 'run4', 'run5']
)
raw_scores_series.fillna(80.0, inplace=True) 
filtered_scores = raw_scores_series[raw_scores_series <= 100] 
transformed_scores = filtered_scores + 2.0 
print(transformed_scores)
```
