# Arithmetic Operations

```python
import pandas as pd
```

```python
var = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
print(var)
```

```python
var["C"] = var["A"] + var["B"]
print(var)
```

```python
var["C"] = var["A"] - var["B"]
print(var)
```

```python
var["C"] = var["A"] * var["B"]
print(var)
```

```python
var["C"] = var["A"] / var["B"]
print(var)
```

```python
var["C"] = var["A"] // var["B"]
print(var)
```

```python
var["C"] = var["A"] % var["B"]
print(var)
```

```python
var["Checker"] = var["A"] > var["B"]
var["Check_greater_1"] = (var["A"] > 1) & (var["B"] > 1)
var["Inverted_Checker"] = ~var["Checker"]
print(var)
```
