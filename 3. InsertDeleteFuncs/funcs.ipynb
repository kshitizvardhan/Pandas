# Insert

```python
import pandas as pd
```

```python
df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
print(df)
```

```python
df.insert(2, "C", [7, 8, 9]) # Insert column C at index 2 with values [7, 8, 9]
print(df)
df.insert(2, "D", [10, 11, 12]) # Insert column D at index 2 with values [10, 11, 12]
print(df)
# Insert row at index 1 with values [16, 17, 18, 19]
df.loc[3] = [16, 17, 18, 19] 
print(df)
df.loc[2] = [13, 14, 15, 16] # Insert row at index 2 with values [13, 14, 15, 16]
print(df)
```

```python
df2 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
print(df2)
```

```python
df2["C"] = df2["A"][:2]
print(df2)
df2["C"] = df2[:2]["A"]
print(df2)
# The above two lines do the same thing, but the second line is more efficient because it avoids the need to access the "A" column twice. 
# By slicing the DataFrame first and then accessing the "A" column, we reduce the number of operations and improve performance.
```

# Delete

```python
df2.pop("C") # Remove column C
print(df2)
df2.drop(1, inplace=True) # Remove row with index 1
print(df2)
del df2["A"] # Remove column A
print(df2)
```
