# Differences between CSV and XLS(excel) file formats: 

- CSV format is a plain text format in which values are separated by commas (Comma Separated Values).
- XLS file format is an Excel Sheets binary file format which holds information about all the worksheets in a file, including both content and formatting

### Write CSV

```python
import pandas as pd
```

```python
dict = {"A": [1, 2, 3], "B": [4, 5, 6]}
df = pd.DataFrame(dict)

df.to_csv("output.csv", index_label="Index")
df.to_csv("output_no_index.csv", index=False)
df.to_csv("output_no_header.csv", header=False)
df.to_csv("output_no_index_no_header.csv", index=False, header=False)
df.to_csv("output_tab_delimited.csv", sep="\t")
df.to_csv("output_semicolon_delimited.csv", sep=";")
df.to_csv("output_pipe_delimited.csv", sep="|")
df.to_csv("output_custom_index_custom_header.csv", index_label="Index", header=["Column1", "Column2"])
```

### Read CSV

```python
csv_1 = pd.read_csv("output_no_index.csv")
print("\n")
print(csv_1)
csv_2 = pd.read_csv("output_no_index.csv", nrows=1)
print("\n")
print(csv_2)
csv_3 = pd.read_csv("output_no_index.csv", usecols=["A"])
print("\n")
print(csv_3)
```
